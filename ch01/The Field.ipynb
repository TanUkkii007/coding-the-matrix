```python
def plot_complex(cx, xlim=[0, 4], ylim=[0, 4]):
    plt.xlim(xlim[0], xlim[1])
    plt.ylim(ylim[0], ylim[1])
    plt.scatter([c.real for c in cx], [c.imag for c in cx])
```

```python
# 1.4.3

plot_complex({1 + 2j + z for z in S})
```

```python
#  1.4.7

plot_complex({z*0.5 for z in S})
```

```python
plot_complex({-1*z for z in S}, xlim=[-4,0], ylim=[-4,0])
```

```python
plot_complex({1j*z for z in S}, xlim=[-4,0])
```

```python
# 1.4.8
plot_complex({0.5j*z for z in S}, xlim=[-4,0])
```

```python
plot_complex({1j*z + 2 -1j for z in S}, xlim=[-4,4])
```

```python
import cmath
N = 20
wx = [cmath.e**(2j*n*cmath.pi/N) for n in range(N)]
plot_complex(wx, xlim=[-1,1], ylim=[-1,1])
```

```python
from numbers import Number
class One:
    def __add__(self, other): return self if other == 0 else 0
    __sub__ = __add__
    def __mul__(self, other):
        if isinstance(other, Number):
            return 0 if other == 0 else self
        return other
    def __div__(self, other):
        if other == 0: raise ZeroDivisionError
        return self
    _truediv__ = __div__
    def __rdiv__(self, other): return other
    __rtruediv__ = __rdiv__
    __radd__ = __add__
    __rsub__ = __add__
    __rmul__ = __mul__
    #hack to ensure not (one < 1e-16) by ensuring not (one < x) for every x
    def __lt__(self, other): return False
    def __eq__(self, other):
        if isinstance(other, self.__class__) or other==0:
            return other != 0
        else:
            raise TypeError
    def __hash__(self): return 1
    def __str__(self): return 'one'
    __repr__ = __str__
    def __neg__(self): return self
    def __bool__(self): return True
    def __format__(self, format_spec): return format(str(self), format_spec)

one = One()
zero = 0
```

### 本章で登場した 3 つの体の名前を挙げよ。

- 実数体
- 複素数体

### 複素数の共役とは何か? 複素数の絶対値を求めるためにはこれをどう用いるか?

$c = a + bi$の複素共役は$\bar{c} = a - bi$と定義される。a + biの絶対値は$\sqrt{a^2 + b^2}$だが、これは$\sqrt{(a + bi)*(a - bi)} = \sqrt{c*\bar{c}}$と等しい。


### 複素数の足し算はどのように行われるか?

$(a + bi) + (c + di) = (a + c) + (b + d)i$

### 複素数の掛け算はどのように行われるか?

$(a + bi) * (c + di) = (ac - bd) + (bc + ad)i$

### 平行移動は複素数でどのように定義されるか?

$f(z) = z + z_0$

### スケーリングは複素数でどのように定義されるか?

$f(z) = rz$ (rは実数)

### 180 度回転は複素数でどのように定義されるか?

$f(z) = -z$

$\pi$ラジアンの回転は、

$f(z) = ze^{i\pi}$

$e^{i\pi} = -1$より、

$f(z) = -z$

### 90 度回転は複素数でどのように定義されるか?

$f(z) = iz$

$\frac{\pi}{2}$ラジアンの回転は、

$f(z) = ze^{i\pi/2}$

$f(z) = z(e^{i\pi})^{\frac{1}{2}}$

$e^{i\pi} = -1$より、

$f(z) = z\sqrt{-1}$

$f(z) = iz$

### GF (2) の足し算はどのように行われるか?

1 + 0 = 1\\
1 + 1 = 0
$

### GF (2) の掛け算はどのように行われるか?

0 \times 1 = 0\\
